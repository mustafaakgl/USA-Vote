# Political Clustering & Voter Analysis for the Democratic Party (2016 - 2020)

## 1) Introduction
The 2016 U.S. presidential election saw Donald Trump win a historic victory over Hillary Clinton, defying many pollsters’ expectations. The Democratic Party must now analyze the voting patterns, understand what happened, and strategize for 2020.

This report will use data-driven insights to uncover:

* Who supported Trump? (Demographics, socio-economic conditions, voting trends)
* Who supported Democrats? (Key strongholds, demographics, voter behaviors)
* What mistakes were made? (Missed opportunities, messaging gaps, miscalculations)
* Where should we focus next? (Key states, counties, swing areas to win in 2020)


## 2️) Objectives

* Analyze Trump vs. Clinton voters: What economic and social conditions led to their choices?
* Identify pro-Trump and pro-Democratic states, counties, and cities.
* Find the key voter demographic shifts and target groups for 2020.
* Determine which states and swing counties should be prioritized.
* Recommend strategic changes for the Democratic Party’s messaging and policies.

## 3) Data Collection & Preprocessing

These datasets are used:

* Election Results (2016) → County-level Trump vs. Clinton votes.
* Demographics (Census Data) → Race, income, education, age distribution.
* Economic Data → Unemployment, median income, poverty levels.
* Urban vs. Rural Divide → Population density per county.
* Social & Political Attitudes → Survey-based voter preferences.
  
Data Cleaning Steps:

* Removed missing values & standardized socio-economic indicators.
* Merged political data with county-level economic statistics.
* Applied feature engineering (e.g., economic well-being index, housing affordability).


## 4) Exploratory Data Analysis (EDA)
Key Findings Before Clustering:
* Trump performed best in counties with low college education rates & high economic distress.
* Clinton dominated urban, diverse, and highly-educated regions.
* Rural America had high Republican turnout, while Democratic turnout was lower than expected in key states (e.g., Wisconsin, Pennsylvania, Michigan).
* Economic hardship counties were more pro-Trump, while wealthier, highly-educated counties leaned Democratic.
* Clustering method are applied

## 5) Clustering Results & Voter Profiles
After clustering counties based on economic, demographic, and political factors, we identified four major voter segments:

       Cluster 1: Trump’s Rural Base (Republican Strongholds)
* Low population density, highly rural.
* Low college education rates.
* High poverty & unemployment.
* Strongly anti-establishment, skeptical of globalism.
* States: West Virginia, Kentucky, parts of the Midwest, Deep South.

       Cluster 2: Working-Class Swing Voters (Trump Flip States)
* Former Obama voters who switched to Trump.
* Rust Belt & Midwest: Pennsylvania, Michigan, Wisconsin, Ohio.
* Moderate economic distress, but strong cultural conservatism.
* Democrats lost here due to lack of economic messaging & weak rural outreach.
* Solution: Rebuild economic message, focus on union workers.

      Cluster 3: Urban Democratic Strongholds (Clinton Safe Zones)
* High-income, highly educated voters.
* Large minority populations (Black, Latino, Asian).
* Coastal cities, tech hubs, academic centers.
* States: New York, California, Massachusetts, Washington, Illinois.
* Strategy for 2020: Increase turnout among young & minority voters to counter rural Republican gains.

      Cluster 4: Emerging Democratic Counties (Future Growth Areas)
* Sun Belt suburbs, growing metros.
* More diverse & younger populations.
* Flipped in 2020: Arizona, Georgia, Texas metro areas.
* Strategy: Increase outreach to Latinos & suburban moderates.

## 6) Conclusion

2016 showed a major shift in the American electorate. To win in 2020, Democrats must:
* Recover the Midwest working class.
* Increase urban and minority voter turnout.
* Expand into Sun Belt swing states.
* Improve rural outreach & messaging.

#### Adding new features

```python
republican_results['gender_ratio'] = republican_results['female_percent_2014'] / (100 - republican_results['female_percent_2014'])
democrat_results['gender_ratio'] = democrat_results['female_percent_2014'] / (100 - democrat_results['female_percent_2014'])
```

The gender_ratio feature has been added for more meaningful demographic gender analysis.

```python
republican_results['income_to_poverty_ratio'] = republican_results['median_household_income_2009_2013'] / republican_results['persons_below_poverty_level_percent_2009_2013']
democrat_results['income_to_poverty_ratio'] = democrat_results['median_household_income_2009_2013'] / democrat_results['persons_below_poverty_level_percent_2009_2013']
```

* It is a very important metric to more accurately measure the level of economic prosperity in the region.
* Average income alone is not sufficient, because a high income does not guarantee that a region is prosperous.

```python
# Women's Representativeness - women's representativeness in business, adjusted for their proportion of the population
republican_results['female_representation_power'] = republican_results['women_owned_firms_percent_2007'] / republican_results['female_percent_2014']
democrat_results['female_representation_power'] = democrat_results['women_owned_firms_percent_2007'] / democrat_results['female_percent_2014']
```

It is a metric that compares women's economic activity with their proportion in the population.

```python
# Global Integration Score - measures the region's international connectivity
republican_results['global_integration_score'] = republican_results['foreign_born_persons_percent_2009_2013'] * 0.6 + republican_results['language_other_than_english_spoken_at_home_percent_age_5_plus_2009_2013'] * 0.4
democrat_results['global_integration_score'] = democrat_results['foreign_born_persons_percent_2009_2013'] * 0.6 + democrat_results['language_other_than_english_spoken_at_home_percent_age_5_plus_2009_2013'] * 0.4
```

Measures a region's global connectivity and international integration. It is used to measure how open a region is to international interaction.

```python
# Entrepreneurship Rate - enterprises per 1000 inhabitants
republican_results['Entrepreneurship_rate'] = (republican_results['nonemployer_establishments_2013'] / republican_results['population_2014']) * 1000
democrat_results['Entrepreneurship_rate'] = (democrat_results['nonemployer_establishments_2013'] / democrat_results['population_2014']) * 1000
```

It was created to compare the Entrepreneurship Potential of the Regions, to compare the Entrepreneurship Rate and Party Support Rates and to understand the link between Entrepreneurship and Economic Welfare.

```python
republican_results['vote_turnout_rate'] = republican_results['votes'] / republican_results['population_2014'] * 100
democrat_results['vote_turnout_rate'] = (
    democrat_results['votes'].fillna(0) / democrat_results['population_2014'].replace(0, np.nan)
) * 100

```

This code calculates the voter turnout rate in each area to determine the level of voter turnout. High voter turnout rates reflect the participation and activity of local voters, while low rates indicate areas that potentially need targeting.

```python
republican_results['age_index'] = republican_results['persons_65_and_over_percent_2014'] / republican_results['persons_under_18_percent_2014']
democrat_results['age_index'] = democrat_results['persons_65_and_over_percent_2014'] / democrat_results['persons_under_18_percent_2014']
```

This new variable is used to measure age demographics in the region and to understand the balance between the older and younger population. It was added to better understand the demographic structure of the regions and to investigate party support rates.

```python
# Economic Well-being Index - high income and low poverty = high well-being
republican_results['economic_wellbeing_index'] = (
    republican_results['median_household_income_2009_2013'] / 
    (republican_results['persons_below_poverty_level_percent_2009_2013'] + 1)  # +1 to prevent division by zero
) / 1000  # Scaling for more readable values

democrat_results['economic_wellbeing_index'] = (
    democrat_results['median_household_income_2009_2013'] / 
    (democrat_results['persons_below_poverty_level_percent_2009_2013'] + 1) 
) / 1000  
```

This new variable is a welfare metric developed to better measure economic living standards in the region. It was created to more accurately measure economic well-being.

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

republican_results['ses_score'] = (
    normalize(republican_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(republican_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(republican_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)

democrat_results['ses_score'] = (
    normalize(democrat_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(democrat_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(democrat_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)
```

In order to analyse the socioeconomic structure of the regions and observe voter behaviour, the socioeconomic score was created with the education rate, income and home ownership rate.

```python

republican_results['housing_density'] = republican_results['housing_units_2014'] / republican_results['land_area_in_square_miles_2010']

republican_results['housing_affordability_index'] = republican_results['median_household_income_2009_2013'] / republican_results['median_housing_value_owner_occupied_units_2009_2013']

democrat_results['housing_density'] = democrat_results['housing_units_2014'] / democrat_results['land_area_in_square_miles_2010']
democrat_results['housing_affordability_index'] = (democrat_results['median_household_income_2009_2013'] / democrat_results['median_housing_value_owner_occupied_units_2009_2013'])

```

The housing density metric allows to understand how urbanised a region is and to better compare rural and urban areas.
The housing affordability index was created to measure how accessible the housing market is. It measures the ability of people in the region to purchase housing according to their income level.

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

merged_results['ses_score'] = (
    normalize(merged_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(merged_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(merged_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)

county_ses_comparison = merged_results.groupby(['county', 'party'])['ses_score'].median().reset_index()

county_ses_pivot = county_ses_comparison.pivot(index='county', columns='party', values='ses_score')

county_ses_pivot['SES_Difference'] = abs(county_ses_pivot['Democrat'] - county_ses_pivot['Republican'])
top_counties = county_ses_pivot.sort_values(by='SES_Difference', ascending=False).head(20)
plt.figure(figsize=(14, 6))
top_counties[['Democrat', 'Republican']].plot(kind='bar', figsize=(14, 6), colormap="coolwarm")

plt.title('SES Score Comparison by County (Top Differences)')
plt.xlabel('County')
plt.ylabel('Median SES Score')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Party")

# Show plot
plt.show()
```

In most of the 20 districts shown in the graph, Democratic voters have higher socioeconomic statu scores than Republican voters. This suggests that Democrats are attracting voters of generally higher socioeconomic status. This graph reinforces the narrative of ‘two Americas’ - more educated, high-income Democrats on one side, and generally lower SES Republicans on the other. Special attention should be paid to counties where the gap between Democratic and Republican SES scores is small (e.g. Nevada, Mecklenburg, Oldham) - these counties are easier to win.In the Republican-dominated counties outside Forsyth (Platte, Guadalupe, Beaufort), the margin is relatively small. These counties can be won with the right message.

## Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Merge Republican and Democrat datasets
combined_results = pd.concat([republican_results, democrat_results], axis=0, ignore_index=True)

# Define the relevant columns for clustering
cluster_features = [
    'median_household_income_2009_2013', 'per_capita_income_past_12_months_2013_d',
    'persons_below_poverty_level_percent_2009_2013', 'private_nonfarm_employment_2013',
    'Entrepreneurship_rate', 'median_housing_value_owner_occupied_units_2009_2013',
    'housing_affordability_index', 'ses_score'
]

# Drop missing values
cluster_data = combined_results[cluster_features].dropna()

# Standardize the data (Z-Score Normalization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(cluster_data)

# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-Cluster Sum of Squares (WCSS)
K_values = range(1, 10)  # Testing clusters from 1 to 9

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Elbow Method Visualization
plt.figure(figsize=(6, 4))
plt.plot(K_values, wcss, marker='o', linestyle='-')
plt.title("Elbow Method: Finding the Optimal Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Within Cluster Sum of Squares)")
plt.show()

# Define the optimal number of clusters
optimal_k = 4  # Selected based on the Elbow Method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)

# Assign cluster labels to the dataset
cluster_data['cluster'] = kmeans.labels_

# Retrieve cluster centers (in standardized scale)
centroids = kmeans.cluster_centers_

# Convert to DataFrame for better readability
centroids_df = pd.DataFrame(centroids, columns=cluster_features)
print("Cluster Centers (Standardized Data):")
display(centroids_df)

# Summarizing cluster means
summary = cluster_data.groupby("cluster").mean()
print("Mean Values per Cluster (Standardized Scale):")
display(summary)

# Transform cluster centers back to the original scale
original_centroids = scaler.inverse_transform(centroids)
original_centroids_df = pd.DataFrame(original_centroids, columns=cluster_features)
print("Cluster Centers (Original Data Scale):")
display(original_centroids_df)

# PCA for visualization in 2D space
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

cluster_data['pca_1'] = principal_components[:, 0]
cluster_data['pca_2'] = principal_components[:, 1]

# Visualizing Clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='pca_1', 
    y='pca_2', 
    hue='cluster', 
    data=cluster_data, 
    palette='viridis',
    alpha=0.7
)
plt.title("Cluster Distribution (PCA 2D Representation)")
plt.show()
```

Cluster 0: Low Income, High Poverty Regions
* Low income level (median household income and per capita income are low).
* High poverty rate, i.e. a cluster where low-income individuals are concentrated.
* Private sector employment is low and entrepreneurship rate is also low.
* Low home ownership value, probably located in low-income areas.
* Cluster with the lowest socio-economic status score.
* Housing accessibility index is high, indicating that it is easier to buy a house here (because prices may be low).

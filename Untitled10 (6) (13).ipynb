#### Adding new features

```python
republican_results['gender_ratio'] = republican_results['female_percent_2014'] / (100 - republican_results['female_percent_2014'])
democrat_results['gender_ratio'] = democrat_results['female_percent_2014'] / (100 - democrat_results['female_percent_2014'])
```

The ratios of women and men have been determined and used in regional demographic analyses.

```python
republican_results['retail_per_person'] = republican_results['retail_sales_2007'] / republican_results['population_2014']
republican_results['manufacturing_per_person'] = republican_results['manufacturers_shipments_2007'] / republican_results['population_2014']
democrat_results['retail_per_person'] = democrat_results['retail_sales_2007'] / democrat_results['population_2014']
democrat_results['manufacturing_pea_person'] = democrat_results['manufacturers_shipments_2007'] / democrat_results['population_2014']

```

High regional retail spending indicates that the region's consumption power is buoyant, while manufacturing shipments reflect local economic dynamics. These measures are useful for observing the relationship between voter behaviour and the economic situation.

```python
republican_results['income_to_poverty_ratio'] = republican_results['median_household_income_2009_2013'] / republican_results['persons_below_poverty_level_percent_2009_2013']
democrat_results['income_to_poverty_ratio'] = democrat_results['median_household_income_2009_2013'] / democrat_results['persons_below_poverty_level_percent_2009_2013']
```

This feature divides the average household income in each region by the poverty rate, creating a metric called ‘income_to_poverty_ratio’. This ratio indicates the economic prosperity and poverty rate of a region.

```python
# Women's Representativeness - women's representativeness in business, adjusted for their proportion of the population
republican_results['female_representation_power'] = republican_results['women_owned_firms_percent_2007'] / republican_results['female_percent_2014']
democrat_results['female_representation_power'] = democrat_results['women_owned_firms_percent_2007'] / democrat_results['female_percent_2014']
```

It measures the representation of women in the business world by normalising their representation in the general population.

```python
# Global Integration Score - measures the region's international connectivity
republican_results['global_integration_score'] = republican_results['foreign_born_persons_percent_2009_2013'] * 0.6 + republican_results['language_other_than_english_spoken_at_home_percent_age_5_plus_2009_2013'] * 0.4
democrat_results['global_integration_score'] = democrat_results['foreign_born_persons_percent_2009_2013'] * 0.6 + democrat_results['language_other_than_english_spoken_at_home_percent_age_5_plus_2009_2013'] * 0.4
```

Measures a region's global connectivity and international integration. It combines the proportion of foreign-born population and the proportion speaking a language other than English at home.The coefficients 0.6 and 0.4 constitute the total score with more weight given to the foreign-born ratio.

```python
# Entrepreneurship Rate - enterprises per 1000 inhabitants
republican_results['Entrepreneurship_rate'] = (republican_results['nonemployer_establishments_2013'] / republican_results['population_2014']) * 1000
democrat_results['Entrepreneurship_rate'] = (democrat_results['nonemployer_establishments_2013'] / democrat_results['population_2014']) * 1000
```

It shows the number of self-employed or self-employed persons per 1000 inhabitants. It reflects the entrepreneurial spirit and small business ownership of a society.

```python
republican_results['vote_turnout_rate'] = republican_results['votes'] / republican_results['population_2014'] * 100
democrat_results['vote_turnout_rate'] = (
    democrat_results['votes'].fillna(0) / democrat_results['population_2014'].replace(0, np.nan)
) * 100

```

This code calculates the voter turnout rate in each area to determine the level of voter turnout. High voter turnout rates reflect the participation and activity of local voters, while low rates indicate areas that potentially need targeting.

```python
republican_results['age_index'] = republican_results['persons_65_and_over_percent_2014'] / republican_results['persons_under_18_percent_2014']
democrat_results['age_index'] = democrat_results['persons_65_and_over_percent_2014'] / democrat_results['persons_under_18_percent_2014']
```

Summarises the age structure of the state or district in a single value. This index allows one to examine how Trump support varies by age demographics. In particular, one can ask whether Trump support is higher in areas with a high proportion of older people. 

```python
# Economic Well-being Index - high income and low poverty = high well-being
republican_results['economic_wellbeing_index'] = (
    republican_results['median_household_income_2009_2013'] / 
    (republican_results['persons_below_poverty_level_percent_2009_2013'] + 1)  # +1 to prevent division by zero
) / 1000  # Scaling for more readable values

democrat_results['economic_wellbeing_index'] = (
    democrat_results['median_household_income_2009_2013'] / 
    (democrat_results['persons_below_poverty_level_percent_2009_2013'] + 1) 
) / 1000  
```

This index measures the economic health of a region. High income and low poverty rate indicate better economic well-being.

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

republican_results['ses_score'] = (
    normalize(republican_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(republican_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(republican_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)

democrat_results['ses_score'] = (
    normalize(democrat_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(democrat_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(democrat_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)
```

In order to analyse the socioeconomic structure of the regions and observe voter behaviour, the socioeconomic score was created with the education rate, income and home ownership rate.

```python

republican_results['housing_density'] = republican_results['housing_units_2014'] / republican_results['land_area_in_square_miles_2010']

republican_results['housing_affordability_index'] = republican_results['median_household_income_2009_2013'] / republican_results['median_housing_value_owner_occupied_units_2009_2013']

democrat_results['housing_density'] = democrat_results['housing_units_2014'] / democrat_results['land_area_in_square_miles_2010']
democrat_results['housing_affordability_index'] = (democrat_results['median_household_income_2009_2013'] / democrat_results['median_housing_value_owner_occupied_units_2009_2013'])

```

The housing density metric reveals which regions have more densely populated areas and can play a critical role in urban planning, infrastructure investments and local quality of life strategies. The Housing Accessibility Index shows how accessible housing is to local residents. This information can provide important insights into the economic situation, cost of living and housing policies, and can guide public and private sector investment strategies.

```python
def normalize(series):
    return (series - series.min()) / (series.max() - series.min())

merged_results['ses_score'] = (
    normalize(merged_results["bachelor's_degree_or_higher_percent_age_25_plus_2009_2013"]) * 0.4 + 
    normalize(merged_results["median_household_income_2009_2013"]) * 0.4 + 
    normalize(merged_results["median_housing_value_owner_occupied_units_2009_2013"]) * 0.2
)

county_ses_comparison = merged_results.groupby(['county', 'party'])['ses_score'].median().reset_index()

county_ses_pivot = county_ses_comparison.pivot(index='county', columns='party', values='ses_score')

county_ses_pivot['SES_Difference'] = abs(county_ses_pivot['Democrat'] - county_ses_pivot['Republican'])
top_counties = county_ses_pivot.sort_values(by='SES_Difference', ascending=False).head(20)
plt.figure(figsize=(14, 6))
top_counties[['Democrat', 'Republican']].plot(kind='bar', figsize=(14, 6), colormap="coolwarm")

plt.title('SES Score Comparison by County (Top Differences)')
plt.xlabel('County')
plt.ylabel('Median SES Score')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Party")

# Show plot
plt.show()
```

In most of the 20 districts shown in the graph, Democratic voters have higher SES scores than Republican voters. This suggests that Democrats are attracting voters of generally higher socioeconomic status. This graph reinforces the narrative of ‘two Americas’ - more educated, high-income Democrats on one side, and generally lower SES Republicans on the other. Special attention should be paid to counties where the gap between Democratic and Republican SES scores is small (e.g. Nevada, Mecklenburg, Oldham) - these counties are easier to win.In the Republican-dominated counties outside Forsyth (Platte, Guadalupe, Beaufort), the margin is relatively small. These counties can be won with the right message.

```python
merged_results['Entrepreneurship_rate'] = (
    merged_results['nonemployer_establishments_2013'] / merged_results['population_2014']
) * 1000

county_entrepreneurship = merged_results.groupby(['county', 'party'])['Entrepreneurship_rate'].median().reset_index()
county_entrepreneurship_pivot = county_entrepreneurship.pivot(index='county', columns='party', values='Entrepreneurship_rate')
county_entrepreneurship_pivot_cleaned = county_entrepreneurship_pivot.dropna()

county_entrepreneurship_pivot_cleaned['Entrepreneurship_Difference'] = abs(
    county_entrepreneurship_pivot_cleaned['Democrat'] - county_entrepreneurship_pivot_cleaned['Republican']
)
top_counties = county_entrepreneurship_pivot_cleaned.sort_values(by='Entrepreneurship_Difference', ascending=False).head(20)

plt.figure(figsize=(14, 6))
top_counties[['Democrat', 'Republican']].plot(kind='bar', figsize=(14, 6), colormap="coolwarm")

plt.title('Entrepreneurship Rate Comparison by County (Top Differences)')
plt.xlabel('County')
plt.ylabel('Median Entrepreneurship Rate (Businesses per 1000 Inhabitants)')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title="Party")
plt.show()
```

Highly entrepreneurial regions won by Democrats are likely to focus on technology, service and creative industries, while highly entrepreneurial regions won by Republicans are more likely to be traditional, agricultural or small family-run businesses. #### Analysis of Districts Where Republicans Won, but Have Low Education and Income, and High Ethnic Diversity

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

features = [
    'median_household_income_2009_2013',  # Income level
    "bachelor's_degree_or_higher_percent_age_25_plus_2009_2013",  # Education level
    'persons_below_poverty_level_percent_2009_2013',  # Poverty rate
    'black_alone_percent_2014',  # Black population percentage
    'hispanic_latino_percent_2014',  # Hispanic population percentage
    'asian_alone_percent_2014'  # Asian population percentage
]

republican_won = republican_results[republican_results['candidate'].notna()]
republican_won = republican_won[features].dropna()

scaler = StandardScaler()
scaled_data = scaler.fit_transform(republican_won)

optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_data)

republican_won['Republican_Won_Cluster'] = clusters

cluster_summary = republican_won.groupby('Republican_Won_Cluster')[features].mean()

import seaborn as sns
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plot_data = pd.DataFrame({
    'PCA1': pca_data[:, 0],
    'PCA2': pca_data[:, 1],
    'Cluster': clusters
})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=plot_data, x='PCA1', y='PCA2', hue='Cluster', palette='viridis', alpha=0.6)
plt.title('K-Means Clustering Visualization (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
cluster_summary
```

This analysis yielded three distinct clusters of voters:

* Cluster 2: High Income, Highly Educated, But Strong Republican Areas. College Graduate Rate: 32.88% (Highest!) Black Voters: 8.41% Hispanic Voters: 10.61% Asian Voters: 3.95%. When we look at the dataset as a whole, we see that people with higher education levels and minorities generally vote for Democrats.
